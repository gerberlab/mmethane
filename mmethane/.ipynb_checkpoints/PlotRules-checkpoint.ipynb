```python
def convert_to_log_normal(mu,var):
    var2 = np.log(var/(mu**2) + 1)
    mu2 = np.log(mu) - var2/2
    return mu2, var2
```

```python
tau_wts = 1000
tau_thresh = 5
tau_det = 1
tau_rule = 1
bp = {k:v[-1] for k,v in param_dict.items()}
num_rules = 10
num_otu_centers=12
emb_dim = emb_locs.shape[-1]

emb_locs = torch.tensor(emb_locs)
x = torch.tensor(tr_dataset['metabs']['X'].values).unsqueeze(-1)
kappa = bp['metabs_kappa'].detach().exp().unsqueeze(-1)
dist = (bp['metabs_eta'].detach().reshape(num_rules, num_otu_centers, 1, emb_dim) - emb_locs).norm(2, dim=-1)
otu_wts = torch.sigmoid((kappa-dist)*tau_wts)

otu_wts = otu_wts.to(torch.float64)
feat_combos = (torch.einsum('kij,sjt->skit', otu_wts, x) + 1e-10) / (torch.sum(otu_wts,2).unsqueeze(0).unsqueeze(-1) + 1e-10)
det_activations = torch.sigmoid((feat_combos.squeeze(-1)-bp['metabs_thresh'].detach())*tau_thresh)

z_d = binary_concrete(bp['alpha'].detach(), k=1, hard=False,use_noise=False, noise_factor=1)
z_r = binary_concrete(bp['beta'].detach(), k=1, hard=False, use_noise=False, noise_factor=1)

rule_activations = (1 - z_d.mul(1 - det_activations)).prod(dim=-1)

x_out = F.linear(rule_activations, bp['weight'].detach().to(torch.float64) * z_r.unsqueeze(0), bp['bias'].detach().to(torch.float64))
```

```python
ra_dict = {}
rix_dict={}
da_dict={}
dix_dict={}
xdata = tr_dataset['metabs']['X']
for k, r in enumerate(z_r.numpy()):
    if r >0.1:
        ra_dict[k] = rule_activations[:,k].numpy()
        rix_dict[k] = r
        for j, d in enumerate(z_d[k,:].numpy()):
            if k not in da_dict.keys():
                da_dict[k]={}
                dix_dict[k]={}
            if d > 0.1:
                feat_wts = otu_wts.numpy()[k,j,:]
                feats = [xdata[f]*feat_wts[i] for i,f in
                                 enumerate(xdata.columns.values) if feat_wts[i]/sum(feat_wts)>0.1]
                da_dict[k][j]={'activators':det_activations.numpy()[:,k,j], 'dtype':'metabs',
                               'thresh':bp['metabs_thresh'].detach().numpy()[k,j],
                               'radius':bp['metabs_kappa'].detach().numpy()[k,j],
                               'features':pd.concat(feats, axis=1),
                              'y':tr_dataset['metabs']['y']}
                dix_dict[k][j]=d
        
        
```
